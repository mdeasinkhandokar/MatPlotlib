Scatter Plot

```python
import matplotlib .pyplot as plt
days=[1,2,3,4,5,6,7]
numbers=[2,3,1,4,5,3,6]
plt.scatter(days,numbers)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel
plt.show()
```

```python
import matplotlib .pyplot as plt
days=[1,2,3,4,5,6,7]
numbers=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
size=[40,50,60,70,50,40,60]
plt.scatter(days,numbers,c=colors,s=size,alpha=0.5,marker="*")

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel
plt.show()
```

```python
import matplotlib .pyplot as plt
days=[1,2,3,4,5,6,7]
numbers=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
size=[400,500,600,700,500,400,600]
plt.scatter(days,numbers,c=colors,s=size,marker="*",edgecolor="g",linewidth=3)

plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel
plt.show()
```

```python
import matplotlib .pyplot as plt
days=[1,2,3,4,5,6,7]
numbers=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
size=[400,500,600,700,500,400,600]
plt.scatter(days,numbers,c=colors,s=size,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel
plt.show()
```

```python
import matplotlib .pyplot as plt
days=[1,2,3,4,5,6,7]
numbers=[2,3,1,4,5,3,6]
numbers2=[3,2,4,5,8,6,5]
colors=["r","y","g","b","r","g","r"]
size=[400,500,600,700,500,400,600]
plt.scatter(days,numbers,c=colors,s=size,cmap="viridis")
plt.scatter(days,numbers2,c=colors,s=size,cmap="viridis",alpha=0.5)
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Days",fontsize=15)
plt.ylabel
plt.show()
```
