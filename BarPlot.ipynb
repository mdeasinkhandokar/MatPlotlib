Bar Plot


```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]
y=[85,70,60,82]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("WsCube",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.2,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)

plt.show()
```

Label Parameter

```python
import matplotlib.pyplot as plt
x=["python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("WsCube",fontsize=15)
plt.bar(x,y,width=0.4,color="r",label="popularity")
plt.bar(x,z,width =0.4,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("WsCube",fontsize=15)
width=0.2
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p,z,width,color="y",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","C","C++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]
wdith=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.xlabel("Languages",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("WsCube",fontsize=15)
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p1,z,width,color="y",label="popularity1")
plt.xticks(p+width/2,x,rotation =20)
plt.legend()
plt.show()
```
