Stem Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,3,5,6,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",use_line_collection=False)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,3,5,6,3]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",use_line_collection=False,orientation="horizontal")
plt.legend()

plt.show()
```
